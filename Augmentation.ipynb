```
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os
import keras
from keras.preprocessing.image import ImageDataGenerator
%matplotlib inline
```

```
image_path='drive/My Drive/Augmentation /cam/'
```

```
os.listdir(image_path)
```

```
image_lists = os.listdir(image_path)
```

```
il=[]
for i in image_lists[:]:
  temp_arr=plt.imread(image_path+i)
  il.append(temp_arr)
  #plt.imshow(temp_arr)
  #plt.show()
```

```
gen=ImageDataGenerator(rotation_range=360,width_shift_range=[-200,200],height_shift_range=0.3,shear_range=0.15,zoom_range=[0.5,1.5],brightness_range=(0.1, 0.9),channel_shift_range=150.,horizontal_flip=True,vertical_flip=True)
```

```
single_sample = np.expand_dims(il[20], axis=0)
single_sample.shape
```

```
itt = gen.flow(single_sample, batch_size=1)
```

```

from tqdm import tnrange, tqdm_notebook
```
